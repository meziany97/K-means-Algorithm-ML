```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

# import the dataset

```python
df=pd.read_csv("data.csv")
```

# scaler the input data that we will use in the training using MinMaxScaler

```python

scaler=MinMaxScaler()
scaler.fit(df[['Income($)']])
df['Income($)']=scaler.transform(df[['Income($)']])
scaler.fit(df[['Age']])
df['Age']=scaler.transform(df[['Age']])
```

# fitting the model on our data with number of clusters = 10 in this case

```python

wcss=[]
for i in range(1,10):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df[['Age','Income($)']])
    wcss.append(kmeans.inertia_)
```

# plotting the data

```python
plt.plot(range(1,10),wcss)
plt.title("wcss")
plt.xlabel('nb')
plt.show()
```

```python
kmeans=KMeans(n_clusters=3)
pred=kmeans.fit_predict(df[['Age','Income($)']])

df['cluster']=pred
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
```

```python
# plt.scatter(df1.Age,df1['Income($)'],c='red')
# plt.scatter(df2.Age,df2['Income($)'],c='blue')
plt.scatter(df.Age,df['Income($)'],c=pred)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker="*",s=200,c="purple")
plt.show()
```
